```python
#importera nödvängdig biblotek
import pandas as pd
import numpy as np 
import plotly.express as px
import dash

#Gör listor med filernas namn och namn som dictionary kommer ha.
data = ["Data/Gender_data.csv", "Data/Municipality_Weekly_Data.csv","Data/National_Daily_Deaths.csv", 
"Data/Regional_Daily_Cases.csv", "Data/Regional_Totals_Data.csv"]

data_name = ["Gender", "Municipality", "National_daily_deaths", "Regional_daily", "Regional_total" ]

df_dic = {}

#for loop med enumerate av "data". Detta skapar en iterable "i" som uppdateras för varje loop och ger "df_dic" ett specifikt namn för varje varje variant.
for i,d in enumerate(data):
    df_dic[data_name[i]] = pd.read_csv(f"{d}", encoding="ISO-8859-1", header=0)
```

```python
#Visualisering 1

område = ["Blekinge", "Dalarna", "Gotland", "Gävleborg", "Halland", "Jämtland Härjedalen", "Jönköping", "Kalmar"]

smittade_regional = {}

#Gör en till enumerate for loop. Men denna gång så ger jag dictionaty massa tomma listor.
for i,d in enumerate(område):
    smittade_regional[område[i]] = []
    #Här loopas igenom hela filen som är runt 374.
    for z in range(len(df_dic["Regional_daily"])):
    #Lägger till i en dictionaty värdet av en viss possition i filen. Med "iloc" kan du bestämma en viss rad och kolumn.
        smittade_regional[område[i]].append(df_dic["Regional_daily"].iloc[z,i+2])

#En enkel "try" sats som märker om det blir ett error och printar ut ett medelande istället.
while True:
    try:
        valde = str(input("Välj ett område mellan (Blekinge, Dalarna, Gotland, Gävleborg, Halland, Jämtland Härjedalen, Jönköping, Kalmar"))
        #printar ut en viss område som man har själv valt 
        fig = px.line(y=smittade_regional[valde], x=df_dic["Regional_daily"]["Date"], title=f"Smittade per dag i {valde}", labels={"y":"Antal Smittade", "x":"Datum"})
        fig.show()
        break
    #Medelandet som man får ifall just detta error blir.
    except ValueError:
            print("Kolla att du stavade rätt")

```

```python
#Visualisering 2

#Gör två listor. En på kolumn namnen i filen och en på olika titlar.
kolumn = ["Total_Cases", "Total_Deaths"]
titel = ["Procentuell Smittade av Kvinnor och Men", "Procentuell Döda av Kvinnor och Men"]

#loopar igenom pie grafen och gör två olika.
for i in range(2):
    fig = px.pie(df_dic["Gender"], values=kolumn[i], names='Gender', title=titel[i], color='Gender')
    fig.show()
```

```python
#Visualisering 3

#Två listor på två olika filer och specifika kolumner 
df_daily = ["Regional_daily", "National_daily_deaths"]
column1 = ["Sweden_Total_Daily_Cases", "National_Daily_Deaths"]

#En funktion med en for-sats som funkar likt enumerate och väljer ut viss kolumn i de olika filerna. Denna kolumn blir till en lista som får variabeln "daily(x)".
def daily(x):
    lista = []
    for index, row in df_dic[df_daily[x]].iterrows():
        lista.append(row[column1[x]])
    return lista

#Listor med olika titlar till graferna
titel = ["Smittade per dag i Sverige", "Döda dagligen i Sverige"]
ytitle = ["Antal Smittade", "Antal Döda"]

#likdant som förut för två printningar av grafer.
for i in range(2):
    fig = px.line(y= daily(i), x=df_dic[df_daily[i]]["Date"],title=titel[i], labels={"y":ytitle[i], "x":"Datum"})
    fig.show()

```

```python
#Visualisering 4

#Är likadan som förra grafen men är istället nu ett histogram.
fig = px.histogram(df_dic["National_daily_deaths"], y="National_Daily_Deaths", x="Date", nbins=50, 
labels={"National_Daily_Deaths":"Antal Döda", "Date":"Datum"})
fig.show()
```

```python
#Visualisering 5

total_döda = {}
#Lista med all områden.
region = ["Blekinge", "Dalarna", "Gotland", "Gävleborg", "Halland", "Jämtland Härjedalen", "Jönköping", "Kalmar", "Kronoberg", "Norrbotten", "Skåne", "Stockholm", "Sörmland", "Uppsala", "Värmland", "Västerbotten", "Västernorrland", "Västmanland", "Västra Götaland", "Örebro", "Östergötland"]

#Lägger till specifika värden till det specifika namnet.
for i,d in enumerate(region):
    total_döda[region[i]] = df_dic["Regional_total"].iloc[i,4]

#Sorterar dictionary med bara sälja "value" som är värdena i dictionary till en ny variabel.
döda_värde = sorted(total_döda.values(), reverse=True)
sorted_total_döda = {}

#loopar antal i dictionary
for z in döda_värde:
    #loopar nu igen med dict med själva "namnen".
    for k in total_döda.keys():
        #nu ifall den osorterade dict är lika med den sorterade. Ge en tom dict ett nytt värde.
        if total_döda[k] == z:
            sorted_total_döda[k] = total_döda[k]

#För att göra grafen snygg och kunna använda sig av "head(5)".
df_dic["Regional_total"]["Region"] = sorted_total_döda.keys() 
df_dic["Regional_total"]["Total_Cases"] = sorted_total_döda.values() 

fig = px.bar(df_dic["Regional_total"].head(5), x='Region', y='Total_Cases', labels={"Total_Cases":"Totala Antalet Döda","Region":"Område"})
fig.show()
```
